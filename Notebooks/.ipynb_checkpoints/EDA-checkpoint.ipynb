```python
import pandas as pd
```

```python
raw_df = pd.read_csv('../Data/Stellar-Classification-Dataset.csv')
raw_df.insert(0, 'class', raw_df.pop('class'))
raw_df.head()
```

### Let's first check whether there is any missing data

```python
raw_df.isnull().sum()
```

### Check the data types

```python
raw_df.dtypes
```

### Check the distribution of the outcome 

```python
raw_df['class'].value_counts()
```

```python
raw_df['class'].value_counts().plot.bar(color = 'r', width = 0.8)
```

### Taking advice, from [here](https://developers.google.com/machine-learning/data-prep/construct/sampling-splitting/imbalanced-data), we should down-sample. Let's choose 18,500 samples of each type.

```python
galaxy_df = raw_df[raw_df['class'] == 'GALAXY']
galaxy_df = galaxy_df.sample(18500)
```

```python
qso_df = raw_df[raw_df['class'] == 'QSO']
qso_df = qso_df.sample(18500)
```

```python
star_df = raw_df[raw_df['class'] == 'STAR']
star_df = star_df.sample(18500)
```

### Concatenate them back together to obtain one DataFrame

```python
df = pd.concat([galaxy_df, qso_df, star_df])
df = df.reset_index(drop = True)
df.head()
```

```python
len(df)
```

```python
df.to_csv('../Data/Downsampled-Stellar-Classification-Dataset.csv', index = False)
```
